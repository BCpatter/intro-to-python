## Essay Questions 

**Q1**: Describe the difference between the terms **programming** and **computer science**!

Programming is the action of solving mathematical equations using code while computer science is the action of trying to figure out if something is able to be solved by programming. 

**Q2**: Explain what is a **pull request** and elaborate on how this concept fits a *distributed* organization of work!

A pull request is a way for someone to submit changes they've made to a branch of a bigger repository. With this pull equest, the chnages can be accepted and added to the main repository. 

**Q3**: What does **open-source** mean?

Open-source means that the source code of that program is open to the public and the public is free to use and/or change. This allows people to work on the code and program together and improve the software in a way that conforms to their needs. 

**Q4**: What is a *significant* advantage of a "slow" programming language like Python over a faster one like C?

A "slow" program like python allows programmers to write code much more quickly than C because Python manages its own memory.
